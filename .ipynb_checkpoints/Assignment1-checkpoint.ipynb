#### Q1  Observe an image for different quantization levels and sampling rates

```python
from PIL import Image
import PIL.ImageOps as iops
import matplotlib.pyplot as plt
import numpy as np
import sys
```

```python
img1 = Image.open('img1.jpg').convert('L')
img2 = Image.open('img2.jpg').convert('L')
img3 = Image.open('img3.jpg').convert('L')

img = iops.grayscale(img1)
```

### Quantization levels
Using the posterize function from the ImageOps library, we quantize the image to at different levels.

```python
img_q8 = iops.posterize(img, 8)
img_q4 = iops.posterize(img, 4)
img_q2 = iops.posterize(img, 2)
img_q1 = iops.posterize(img, 1)
```

```python
# Utility function for displaying images as a grid.

def img_grid(dpi, margin, spacing, w, h, rows, cols):
    print('ImageDpi = {}, margin = {}, spacing = {}, imageWidth = {}, imageHeight = {}, gridRows = {}, gridCols = {}'.format(dpi, margin, spacing, w, h, rows, cols))
    width = (cols * w + cols * margin + spacing) / dpi
    height = (rows * h + rows * margin + spacing) / dpi

    figsize = (width, height)

    left = margin / dpi / width
    bottom = margin / dpi / height

    wspace = spacing / float(w)
    hspace = spacing / float(h)
    
    return width, height, figsize, left, bottom, wspace, hspace

# Utility function to display an image (or a grid of images) in their original size (without any scaling).

def display_imgs(images, rows, cols, titles=[], spacing=35, margin=50):
    dpi = 72 # in px. DPI of the image. As known via ImageMagick's identify command.
    w, h = 256, 256 # Taking this as a standard size for images, for the sake of the assignment.
    
    width, height, figsize, left, bottom, wspace, hspace = img_grid(dpi, margin, spacing, w, h, rows, cols)

    fig = plt.figure(figsize=figsize, dpi=dpi)

    fig.subplots_adjust(left=left, bottom=bottom, right=1.-left, top=1.-bottom, wspace=wspace, hspace=hspace)
    
    for i in range(rows*cols):
        ax = fig.add_subplot(rows, cols, i+1)
        if titles:
            ax.set_title(titles[i])
        ax.axis('off')
        ax.imshow(np.asarray(images[i]), cmap='gray', vmin=0, vmax=255, interpolation='none')
```

```python
%matplotlib inline

titles = ['8 bit quantization', '4 bit quantization', '2 bit quantization', '1 bit quantization']
images = [img_q8, img_q4, img_q2, img_q1]

display_imgs(images, 2, 2, titles)
```

Quantization primarily governs the number of bits used to store the information regarding the <b>number of colors used in creating the image.</b>

As seen above,

* 8 bit quantization means 2<sup>8</sup> = 256 colors used for the image.
* 4 bit quantization means 2<sup>4</sup> = 16 colors used for the image.
* 2 bit quantization means 2<sup>2</sup> = 4 colors used for the image.
* 1 bit quantization means 2<sup>1</sup> = 2 colors used for the image.

Therefore, generally, an 8 bit quantization level is preferred for images, to have better color range.

### Sampling Rates

```python
img_s11 = img1.resize((128, 128), resample=Image.NEAREST)
img_s12 = img1.resize((128, 128), resample=Image.BILINEAR)

img_s21 = img2.resize((128, 128), resample=Image.NEAREST)
img_s22 = img2.resize((64, 64), resample=Image.NEAREST)
```

```python
%matplotlib inline

images = [img1, img_s11, img_s12, img2, img_s21, img_s22]

titles = ['Original 256x256',
          'Sampled to 128x128\nNearest Neighbour Sampling\nScaled to 256x256',
          'Sampled to 128x128\nBilinear Sampling\nScaled to 256x256',
          'Original 256x256',
          'Sampled to 128x128\nNearest Neighbour Sampling\nScaled to 256x256',
          'Sampled to 64x64\nNearest Neighbour Sampling\nScaled to 256x256',]

display_imgs(images, 2, 3, titles, spacing=52)
```

As evident from above, we observe the an image, when sampled (here, downsampled),

* There's a visible change in the number of pixels used to show the details per square unit of that image.
* Details are lost.
* Image would be better visible if viewed on a smaller display.

Also, there's a visible difference between the sampling done via <b>Nearest Neighbour</b> method and via <b>Bilinear</b> method. Pixels for image sampled using 1st method are much sharper and more visible, detail loss is significant whereas that in case of the 2nd method, the image seems to have blurred. The interpolation is more relative.

#### Q2 Write a program which will transform a given image in such way that the resultant image histogram is equivalent to histogram of another image.  Please transform the same image using gamma transformation for different values of gamma.

```python
# We'll use 2 grayscale images for this question.
img1 = Image.open('img2_grayLOWCON.jpg')
img2 = Image.open('img1_gray.jpg')

img1 = iops.grayscale(img1)
img2 = iops.grayscale(img2)

samp_img = np.asarray(img1)
ref_img = np.asarray(img2)


titles = ['Sample', 'Reference (Specified)']
images = [img1, img2]

display_imgs(images, 1, 2, titles=titles, spacing=53)
```

```python
def histogram(img):
    rows = img.shape[0]
    cols = img.shape[1]

    hist = np.zeros((256))
    
    for i in np.arange(rows):
        for j in np.arange(cols):
#             print('{} {}'.format(i, j))
            a = img.item(i, j)
            hist[a] += 1      
    return hist

def cumulative_hist(hist):
    cum_hist = hist.copy()
    
    for i in np.arange(1, 256):
        cum_hist[i] = cum_hist[i-1] + cum_hist[i]
    return cum_hist
```

```python
def hist_match(samp_img, ref_img):
    rows1 = samp_img.shape[0]
    cols1 = samp_img.shape[1]

    rows2 = ref_img.shape[0]
    cols2 = ref_img.shape[1]

    hist = histogram(samp_img)
    ref_hist = histogram(ref_img)

    cum_hist = cumulative_hist(hist)
    cum_hist_ref = cumulative_hist(ref_hist)

    prob_cum_hist = cum_hist / (rows1*cols1)
    prob_cum_hist_ref = cum_hist_ref / (rows2*cols2)


    K = 256

    new_values = np.zeros((K))

    for a in range(K):
        j = K - 1

        while j >= 0:
            new_values[a] = j
            if j < 0 or prob_cum_hist[a] > prob_cum_hist_ref[j]:
                break
            j = j - 1

    matched = samp_img.copy()

    for i in range(rows1):
        for j in range(cols1):
            a = matched.item(i, j)
            b = new_values[a]
            matched.itemset((i, j), b)

    return matched
```

```python
%matplotlib inline

matched = hist_match(samp_img, ref_img)

hist = histogram(samp_img)
ref_hist = histogram(ref_img)
matched_hist = histogram(matched)

cum_hist = cumulative_hist(hist)
cum_hist_ref = cumulative_hist(ref_hist)
cum_hist_matched = cumulative_hist(matched_hist)

# im = Image.fromarray(np.uint8(plt.cm.gist_earth(matched)*255))
titles = ['Sample', 'Reference', 'Histogram Matched']

images = [samp_img, ref_img, matched]

display_imgs(images, 1, 3, titles)
```
